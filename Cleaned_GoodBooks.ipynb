```python
import pandas as pd
```

```python
df = pd.read_csv("books.csv", encoding="latin1", on_bad_lines="skip")
df
```

```python
print(df.isnull().sum())
```

```python
df.columns = df.columns.str.strip()
```

```python
df.fillna({
    'average_rating': df['average_rating'].median(),  # Fill with median
    'isbn13': df['isbn13'].mode()[0],  # Fill with most frequent value (mode)
    'num_pages': df['num_pages'].median()  # Fill with median
}, inplace=True)
```

```python
df = df.dropna(subset=['publication_date'])
```

```python
df['publication_date'] = pd.to_datetime(df['publication_date'], errors='coerce')
```

```python
df['publication_year'] = df['publication_date'].dt.year
df['publication_month'] = df['publication_date'].dt.month
df['publication_day'] = df['publication_date'].dt.day
```

```python
df['title'] = df['title'].str.strip()
df['authors'] = df['authors'].str.strip()
df['publisher'] = df['publisher'].str.strip()
```

```python
df['title'] = df['title'].str.title()
df['authors'] = df['authors'].str.title()
df['publisher'] = df['publisher'].str.title()
```

```python
df.fillna({'publisher':"Unknown"}, inplace=True)
df.fillna({'authors':"Unknown"}, inplace=True)
```

```python
df.to_csv("cleaned_goodreads_books.csv", index=False)
```
